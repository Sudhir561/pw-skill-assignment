```python
1. Discuss String Slicing and Provide Example?


String slicing in Python is a technique that allows you to extract a portion (or "slice") of a string by specifying a range of indices. 
This is useful for manipulating and accessing substrings within a larger string.

Syntax of String Slicing
The basic syntax for slicing a string is-

string[start:end:step]

start: The index where the slice begins (inclusive).
end: The index where the slice ends (exclusive).
step: The interval or step size between indices (optional).

a) Basic Slicing:
    
text = "Hello, World!"
sliced_text = text[0:5]
print(sliced_text)  # Output: Hello

# The slice [0:5] extracts characters starting from index 0 up to (but not including) index 5. The result is "Hello".


b)Slicing with Omitted start and end:
    
text = "Python Programming"
print(text[:6])   # Output: Python
print(text[7:])   # Output: Programming

# [:6] slices from the start of the string to index 5.
# [7:] slices from index 7 to the end of the string.

c)Slicing with Negative Indices:

text = "Python Programming"
sliced_text = text[-11:-6]
print(sliced_text)  # Output: Progr

# Negative indices count from the end of the string. -11 corresponds to the 11th character from the end (which is "P"), 
# and -6 is the 6th character from the end (which is "g")

d)Slicing with a Step:

text = "abcdef"
sliced_text = text[0:6:2]
print(sliced_text)  # Output: ace

# The slice [0:6:2] starts at index 0, ends at index 6 (exclusive), and selects every second character (step of 2). The result is "ace".

e)Reversing a String with Slicing:
    
 text = "Python"
reversed_text = text[::-1]
print(reversed_text)  # Output: nohtyP


# The slice [::-1] reverses the string by starting at the end and moving backward with a step of -1.   

f)Reversing a String with Slicing:


text = "Python"
reversed_text = text[::-1]
print(reversed_text)  # Output: nohtyP


# The slice [::-1] reverses the string by starting at the end and moving backward with a step of -1.

g)Extracting Substrings from the Middle:


text = "I love programming in Python"
sliced_text = text[7:18]
print(sliced_text)  # Output: programming


# The slice [7:18] extracts characters starting from index 7 (inclusive) up to index 18 (exclusive), resulting in the substring "programming".

h)Using Slice to Skip Characters:


text = "abcdefghijklmnopqrstuvwxyz"
sliced_text = text[::3]
print(sliced_text)  # Output: adgjmpsvy


# The slice [::3] extracts every third character from the string, starting from the beginning.
```

```python
2. Explain the key features of list in python?

Lists in Python are one of the most versatile and widely used data structures. 
They allow you to store and manipulate an ordered collection of items (elements), 
which can be of different data types (integers, strings, objects, etc.). 
Here are the key features of lists in Python:
    
a). Ordered
Lists maintain the order of elements. This means that elements are stored in a specific sequence, and this sequence is preserved. 
For example, if you add elements [1, 2, 3] to a list, they will remain in that order.

Example:

my_list = [10, 20, 30]
print(my_list)  # Output: [10, 20, 30]

b)Mutable
Lists are mutable, meaning that you can change their content after creation. You can add, remove, or modify elements within a list.

Example:

my_list = [1, 2, 3]
my_list[1] = 10  # Modify the second element
print(my_list)   # Output: [1, 10, 3]

c)Heterogeneous Elements
A single list can contain elements of different data types. This makes lists flexible and allows them to store complex data structures.

Example:

my_list = [1, "Hello", 3.14, True]
print(my_list)  # Output: [1, 'Hello', 3.14, True]

d)Dynamic Size
Lists in Python can dynamically grow or shrink in size. You don't need to declare the size of a list upfront, and you can add or 
remove elements as needed.

Example:

my_list = [1, 2, 3]
my_list.append(4)  # Adding an element
print(my_list)     # Output: [1, 2, 3, 4]
my_list.pop()      # Removing the last element
print(my_list)     # Output: [1, 2, 3]

e)Supports Indexing and Slicing
Lists support indexing and slicing, allowing you to access or modify elements at specific positions, as well as extract sublists.

Example:

my_list = [10, 20, 30, 40, 50]
print(my_list[1])    # Output: 20 (accessing an element)
print(my_list[-1])   # Output: 50 (accessing from the end)
print(my_list[1:4])  # Output: [20, 30, 40] (slicing)

f)Supports Various Operations
Lists support a variety of operations, such as concatenation, repetition, membership testing, and built-in functions 
like len(), min(), max(), sum(), and more.

Example:

my_list = [1, 2, 3]
another_list = [4, 5, 6]
combined_list = my_list + another_list  # Concatenation
print(combined_list)  # Output: [1, 2, 3, 4, 5, 6]

repeated_list = my_list * 2  # Repetition
print(repeated_list)  # Output: [1, 2, 3, 1, 2, 3]

print(2 in my_list)  # Membership testing: Output: True

g)Versatile Methods
Python lists come with a wide range of built-in methods to perform different operations like adding, removing, sorting, and 
reversing elements.
Example:


my_list = [3, 1, 4, 1, 5]
my_list.append(9)   # Adds an element to the end
my_list.remove(1)   # Removes the first occurrence of the element
my_list.sort()      # Sorts the list in ascending order
my_list.reverse()   # Reverses the list order
print(my_list)      # Output: [9, 5, 4, 3, 1]

h)Supports Iteration
Lists are iterable, meaning you can loop through them using loops like for or comprehensions to process elements.
Example:

my_list = [1, 2, 3, 4, 5]
for item in my_list:
     print(item)      
# output: 1,2,3,4,5

i)Nested Lists
Lists can contain other lists as elements, allowing you to create multi-dimensional lists (lists of lists).
Example:


nested_list = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
print(nested_list[1])       # Output: [4, 5, 6] (accessing a sublist)
print(nested_list[1][2])    # Output: 6 (accessing an element in a sublist)

j) Memory Efficiency

Python lists are implemented as dynamic arrays, meaning they offer memory efficiency by allocating additional space when needed, 
making them a good choice for dynamic data storage.
```

```python
3)Describe how to access ,modify and delete elements in a list with examples?

In Python, lists are mutable, which means you can access, modify, and delete elements. 
Below are detailed explanations with examples of how to perform these operations.

a). Accessing Elements in a List
You can access elements in a list using indexing. Python uses zero-based indexing, so the first element has an index of 0.

Example:

# Define a list
my_list = [10, 20, 30, 40, 50]

# Access the first element (index 0)
print(my_list[0])  # Output: 10

# Access the third element (index 2)
print(my_list[2])  # Output: 30

# Access the last element using negative indexing
print(my_list[-1])  # Output: 50

b)Modifying Elements in a List
You can modify elements in a list by assigning a new value to a specific index.

Example:

# Define a list
my_list = [10, 20, 30, 40, 50]

# Modify the second element (index 1)
my_list[1] = 25

# Modify the last element using negative indexing
my_list[-1] = 55

print(my_list)  # Output: [10, 25, 30, 40, 55]

c) Modifying Multiple Elements Using Slicing

You can also modify multiple elements in a list using slicing.
Example:

# Define a list
my_list = [10, 20, 30, 40, 50]

# Modify the first three elements
my_list[0:3] = [15, 25, 35]

print(my_list)  # Output: [15, 25, 35, 40, 50]

d)Deleting Elements from a List

You can delete elements from a list using the del statement, remove() method, or pop() method.

 Using the del Statement
The del statement can be used to delete an element at a specific index or delete a slice of elements.

* Delete a single element

# Define a list
my_list = [10, 20, 30, 40, 50]

# Delete the second element (index 1)
del my_list[1]

print(my_list)  # Output: [10, 30, 40, 50]

* Delete multiple elements using slicing


# Define a list
my_list = [10, 20, 30, 40, 50]

# Delete the first three elements
del my_list[0:3]

print(my_list)  # Output: [40, 50]

Using the remove() Method

The remove() method removes the first occurrence of a specific value from the list.

Example:


# Define a list
my_list = [10, 20, 30, 20, 40]

# Remove the first occurrence of 20
my_list.remove(20)

print(my_list)  # Output: [10, 30, 20, 40]

Using the pop() Method

The pop() method removes and returns the element at a specific index. If no index is provided, it removes and returns the last element.

* Remove the element at a specific index


# Define a list
my_list = [10, 20, 30, 40, 50]

# Remove the element at index 2 and return it
removed_element = my_list.pop(2)

print(removed_element)  # Output: 30
print(my_list)          # Output: [10, 20, 40, 50]

* Remove the last element


# Define a list
my_list = [10, 20, 30, 40, 50]

# Remove the last element and return it
removed_element = my_list.pop()

print(removed_element)  # Output: 50
print(my_list)          # Output: [10, 20, 30, 40]

e) Clearing a List

If we want to remove all elements from a list, you can use the clear() method.

Example:

# Define a list
my_list = [10, 20, 30, 40, 50]

# Clear the entire list
my_list.clear()

print(my_list)  # Output: []

Access: Use indexing to access individual elements or slicing to access a range of elements.
Modify: Assign new values to specific indices or use slicing to modify multiple elements.
Delete: Use del, remove(), or pop() to remove elements. Use clear() to remove all elements from the list.
```

```python
4) Compare and contrast tuples and lists with examples?

Tuples and lists are both sequence data types in Python that allow you to store collections of items.
However, they differ in several key aspects, such as mutability, syntax, and use cases. Below is a comparison of tuples and lists 
with examples.

a). Mutability
Lists: Lists are mutable, meaning you can modify, add, or remove elements after the list has been created.
Tuples: Tuples are immutable, meaning once a tuple is created, its elements cannot be changed, added, or removed.

Example:


# List (Mutable)
my_list = [1, 2, 3]
my_list[1] = 10  # Modify an element
print(my_list)   # Output: [1, 10, 3]

# Tuple (Immutable)
my_tuple = (1, 2, 3)
# my_tuple[1] = 10  # This will raise a TypeError: 'tuple' object does not support item assignment

b). Syntax
Lists: Lists are defined using square brackets [].
Tuples: Tuples are defined using parentheses ().
Example:


my_list = [1, 2, 3]  # List
my_tuple = (1, 2, 3)  # Tuple
Note: A tuple with a single element requires a trailing comma, e.g., (1,).

c). Use Cases

Lists: Lists are preferred when you need a collection that is likely to change. 
They are ideal for scenarios where elements need to be added, removed, or modified.

Tuples: Tuples are preferred when you need a collection that should remain constant. 
They are often used to represent fixed collections of items, such as coordinates, or 
when you want to ensure data integrity by preventing modification.

Example:


# List: Shopping cart items (can change)
shopping_cart = ["apple", "banana", "cherry"]
shopping_cart.append("orange")  # Adding an item
print(shopping_cart)  # Output: ['apple', 'banana', 'cherry', 'orange']

# Tuple: Coordinates (should not change)
coordinates = (10, 20)
# coordinates[0] = 15  # This will raise a TypeError
d). Performance
Lists: Lists are generally slower than tuples in terms of performance because of their mutability. 
The overhead of managing the ability to change the list's contents can make operations slightly slower.
Tuples: Tuples are faster than lists because they are immutable. Python can optimize memory and 
speed when working with tuples, making them more efficient for read-only operations.

Example:


import timeit

# Measure the performance of creating a list
list_time = timeit.timeit("[1, 2, 3, 4, 5]", number=1000000)

# Measure the performance of creating a tuple
tuple_time = timeit.timeit("(1, 2, 3, 4, 5)", number=1000000)

print(f"List creation time: {list_time}")
print(f"Tuple creation time: {tuple_time}")
e). Methods
Lists: Lists have a wide range of built-in methods for modifying the list, such as append(), remove(), pop(), extend(), insert(), clear(), 
sort(), and reverse().

Tuples: Tuples have very few built-in methods because they are immutable. The main methods are count() (to count occurrences of an element) 
and index() (to find the index of an element).

Example:


# List methods
my_list = [1, 2, 3]
my_list.append(4)  # Adding an element
print(my_list)  # Output: [1, 2, 3, 4]

# Tuple methods
my_tuple = (1, 2, 3, 2)
print(my_tuple.count(2))  # Output: 2 (count occurrences of 2)
print(my_tuple.index(3))  # Output: 2 (index of element 3)

f). Memory Efficiency
Lists: Lists consume more memory because they are dynamic and allow modifications. 
Python allocates extra memory to lists to accommodate future additions.

Tuples: Tuples consume less memory because they are static and immutable, and Python can optimize their storage.
7. Immutability and Hashability

Lists: Lists are not hashable, which means they cannot be used as keys in a dictionary or elements in a set.
Tuples: Tuples are hashable (as long as all their elements are hashable), which means they can be used as keys in dictionaries 
or elements in sets.

Example:


# Dictionary with tuple as a key
my_dict = {("John", "Doe"): 25}
print(my_dict)  # Output: {('John', 'Doe'): 25}

# Trying to use a list as a key will raise a TypeError
# my_dict = {[1, 2]: "value"}  # TypeError: unhashable type: 'list'

g). Unpacking

Both lists and tuples support unpacking, allowing you to assign the elements to variables directly.
Example:


# List unpacking
my_list = [1, 2, 3]
a, b, c = my_list
print(a, b, c)  # Output: 1 2 3

# Tuple unpacking
my_tuple = (4, 5, 6)
x, y, z = my_tuple
print(x, y, z)  # Output: 4 5 6

h)Flexibility

Lists: Lists offer more flexibility since they can be modified. They are suitable for tasks where the data needs to change over time.

Tuples: Tuples offer more stability due to their immutability. They are suitable for tasks where the data should remain constant.

Summary Table

Feature	                                                   Lists	                                                   Tuples

Mutability                                      	Mutable (can be modified)	                        Immutable (cannot be modified)

Syntax	                                            Square brackets []	                                 Parentheses ()

Methods	                                            Many built-in methods	                             Few methods (e.g., count, index)

Performance                                       	Slower, more overhead	                             Faster, less overhead

Memory	                                            Consumes more memory	                             Consumes less memory

Use Case	                                        Dynamic, changing data	                             Fixed, constant data

Hashability	                                        Not hashable	                                     Hashable

Indexing/Slicing	                                Supported	                                         Supported

Use Lists: When you need a collection that will change over time, such as a shopping cart, to-do list, or dynamic dataset.

Use Tuples: When you need a fixed collection of items that should remain constant, such as geographic coordinates, configuration settings,
or when you want to ensure that data is not modified accidentally.
```

```python
5. Describe the key features  of sets and provide examples  of their use?

Sets in Python are a collection data type that is unordered, mutable, and does not allow duplicate elements.
They are useful for storing and manipulating unique items and performing mathematical set operations 
like union, intersection, and difference. Below are the key features of sets along with examples of their use.

Key Features of Sets in Python:
    
a). Unordered Collection
Sets do not maintain any specific order of elements. When you print a set, the elements may appear in a different order
than they were added.

Since sets are unordered, they do not support indexing, slicing, or other sequence-like behavior.
Example:

my_set = {1, 2, 3, 4}
print(my_set)  # Output: {1, 2, 3, 4} (order may vary)

b). No Duplicate Elements
Sets automatically remove duplicate elements. If you add the same element multiple times, it will only be stored once.
Example:

my_set = {1, 2, 2, 3, 4, 4}
print(my_set)  # Output: {1, 2, 3, 4} (duplicates removed)

c). Mutable
Sets are mutable, meaning you can add or remove elements after the set has been created. 
However, the elements themselves must be immutable (e.g., numbers, strings, tuples).

Example:

my_set = {1, 2, 3}
my_set.add(4)  # Adding an element
print(my_set)  # Output: {1, 2, 3, 4}

my_set.remove(2)  # Removing an element
print(my_set)     # Output: {1, 3, 4}

d). No Indexing or Slicing
Since sets are unordered, they do not support indexing or slicing like lists or tuples. You cannot access elements by their position.

Example:

my_set = {1, 2, 3}
# print(my_set[0])  # This will raise a TypeError: 'set' object is not subscriptable

e). Supports Mathematical Set Operations
Sets provide built-in methods for performing mathematical operations like union, intersection, difference, and symmetric difference.

Example:

set1 = {1, 2, 3}
set2 = {3, 4, 5}

# Union (elements from both sets)
union_set = set1.union(set2)
print(union_set)  # Output: {1, 2, 3, 4, 5}

# Intersection (common elements)
intersection_set = set1.intersection(set2)
print(intersection_set)  # Output: {3}

# Difference (elements in set1 but not in set2)
difference_set = set1.difference(set2)
print(difference_set)  # Output: {1, 2}

# Symmetric Difference (elements in either set, but not both)
symmetric_difference_set = set1.symmetric_difference(set2)
print(symmetric_difference_set)  # Output: {1, 2, 4, 5}

f). Efficient Membership Testing
Sets are highly efficient for checking whether an element is present in the set or not. 
This is due to the underlying implementation using hash tables.

Example:

my_set = {1, 2, 3, 4, 5}

# Check if 3 is in the set
print(3 in my_set)  # Output: True

# Check if 6 is not in the set
print(6 not in my_set)  # Output: True

g). Set Comprehension
Just like list comprehensions, Python supports set comprehensions to create sets based on existing iterables.
Example:


# Create a set of squares of numbers from 1 to 5
squares_set = {x**2 for x in range(1, 6)}
print(squares_set)  # Output: {1, 4, 9, 16, 25}

h). Immutable Counterpart: frozenset
Python also provides an immutable version of a set called frozenset. Once created, you cannot modify the elements of a frozenset.

Example:

my_frozenset = frozenset([1, 2, 3, 4])
print(my_frozenset)  # Output: frozenset({1, 2, 3, 4})

# my_frozenset.add(5)  # This will raise an AttributeError: 'frozenset' object has no attribute 'add'

i). Set Methods
Sets come with various built-in methods to manipulate their elements, including add(), remove(), discard(), pop(), clear(), copy(), and more.
Example:

my_set = {1, 2, 3}

my_set.add(4)  # Add an element
my_set.discard(2)  # Remove an element (no error if element doesn't exist)
popped_element = my_set.pop()  # Remove and return an arbitrary element
print(my_set)  # Output: {3, 4} (after popping an element)

*Use Cases for Sets:
Removing Duplicates from a List

You can convert a list to a set to remove duplicate elements, as sets do not allow duplicates.
Example:

my_list = [1, 2, 2, 3, 4, 4, 5]
unique_elements = set(my_list)
print(unique_elements)  # Output: {1, 2, 3, 4, 5}

*Membership Testing

Sets are ideal for fast membership testing, such as checking if an element is present in a collection.
Example:

allowed_users = {"alice", "bob", "carol"}
print("alice" in allowed_users)  # Output: True
Mathematical Set Operations

Sets are perfect for performing mathematical operations like finding common elements (intersection) or combining elements from two collections (union).
Example:

students_in_class_A = {"john", "alice", "mike"}
students_in_class_B = {"mike", "sarah", "alice"}

# Students who are in both classes
common_students = students_in_class_A.intersection(students_in_class_B)
print(common_students)  # Output: {'mike', 'alice'}
Eliminating Duplicates While Maintaining Order



Sets in Python are a powerful and efficient data structure for storing unique items and performing set operations. 
They are particularly useful when you need to eliminate duplicates, perform fast membership testing, and 
conduct mathematical operations on collections of items. Their simplicity and flexibility make them a valuable tool in 
many programming scenarios.
```

```python
7.describe how to add ,modify and delete items in dictionary with examples?

Dictionaries in Python are collections of key-value pairs that are mutable, meaning you can add, modify, and 
delete items after the dictionary has been created. Below is a description of how to perform these operations with examples.

a). Adding Items to a Dictionary
You can add items to a dictionary by assigning a value to a new key. If the key does not already exist in the dictionary, 
it will be added along with the corresponding value.

Example:

# Create a dictionary
my_dict = {'name': 'Alice', 'age': 25}

# Add a new key-value pair
my_dict['city'] = 'New York'

print(my_dict)  # Output: {'name': 'Alice', 'age': 25, 'city': 'New York'}

b). Modifying Items in a Dictionary

To modify an item in a dictionary, you can assign a new value to an existing key. If the key already exists, its value will be updated.

Example:

# Modify the value of an existing key
my_dict['age'] = 26

print(my_dict)  # Output: {'name': 'Alice', 'age': 26, 'city': 'New York'}

You can also use the update() method to modify or add multiple items at once.

Example:

# Modify multiple items using the update() method
my_dict.update({'age': 27, 'city': 'Los Angeles'})

print(my_dict)  # Output: {'name': 'Alice', 'age': 27, 'city': 'Los Angeles'}

c). Deleting Items from a Dictionary
You can delete items from a dictionary using the del statement, the pop() method, or the popitem() method.

a. Using del Statement

The del statement removes a key-value pair by specifying the key.

Example:

* Delete a key-value pair using del

del my_dict['city']

print(my_dict)  # Output: {'name': 'Alice', 'age': 27}

*. Using pop() Method

The pop() method removes a key-value pair and returns the value associated with the key.

Example:

# Delete a key-value pair using pop and capture the removed value

removed_value = my_dict.pop('age')

print(my_dict)  # Output: {'name': 'Alice'}
print(removed_value)  # Output: 27

*. Using popitem() Method

The popitem() method removes and returns the last inserted key-value pair. This method is useful in scenarios where the order of insertion 
matters (Python 3.7+).

Example:

# Add some items back for demonstration
my_dict['age'] = 27
my_dict['city'] = 'Los Angeles'

# Delete the last inserted key-value pair using popitem
removed_pair = my_dict.popitem()

print(my_dict)  # Output: {'name': 'Alice', 'age': 27}
print(removed_pair)  # Output: ('city', 'Los Angeles')

*. Using clear() Method

The clear() method removes all items from the dictionary, leaving it empty.

Example:

# Clear all items from the dictionary
my_dict.clear()

print(my_dict)  # Output: {}

Summary:
    
Adding Items: You can add items by assigning a value to a new key, or using the update() method to add multiple items at once.

Modifying Items: You can modify an item by assigning a new value to an existing key, or using the update() method to modify multiple items.

Deleting Items: You can delete items using the del statement, pop() method, popitem() method, or clear() method for removing all items.
```

```python
8. Discuss the importance of dictionary keys being immutable and provide examples

Importance of Dictionary Keys Being Immutable in Python

In Python, dictionary keys must be immutable. This is because dictionaries use a hashing mechanism to store and retrieve values efficiently. For a key to be hashable (i.e., to have a consistent hash value), it must be immutable. Immutable objects cannot change once created, ensuring that the hash value remains constant, which is essential for the performance and integrity of the dictionary.

Why Immutability of Keys Is Important-

Consistent Hashing:

Dictionaries in Python are implemented as hash tables. When you add a key-value pair to a dictionary, 
the key is passed through a hash function that generates a unique hash value. This hash value is used to determine where the value
will be stored in memory.

If a key were mutable, changing its value after it was added to the dictionary would change its hash value, leading to inconsistency. 
The dictionary would no longer be able to locate the key's associated value, resulting in errors.

Efficient Lookups:

Since dictionaries rely on hash tables, lookups (checking if a key exists, or retrieving a value by key) are very efficient,
typically O(1) in time complexity. This efficiency depends on the immutability of keys because the hash function needs 
to produce consistent results for the same key.

If keys could change, the hash table would need to rehash keys each time they changed, which would significantly reduce performance.
Dictionary Integrity:

The immutability of keys also ensures the integrity of the dictionary. If keys could change, it could lead to unpredictable behavior. 

For example, a key might suddenly "disappear" from the dictionary if its value changes and the hash table can no longer locate it.
Examples of Immutable vs. Mutable Objects as Keys

Immutable Objects: These can be used as dictionary keys because they are hashable. Examples include integers, strings, tuples
(if they contain only other immutable objects), and frozensets.

Mutable Objects: These cannot be used as dictionary keys because their hash value can change. 

Examples include lists, sets, and dictionaries.

Examples

a. Using Immutable Keys

# Integer as key (immutable)
my_dict = {1: "one", 2: "two"}

# String as key (immutable)
my_dict["name"] = "Alice"

# Tuple as key (immutable)
my_dict[(1, 2)] = "coordinates"

print(my_dict)
# Output: {1: 'one', 2: 'two', 'name': 'Alice', (1, 2): 'coordinates'}

b. Using Mutable Keys (Not Allowed)

# List as key (mutable) - This will raise a TypeError
try:
    my_dict = {[1, 2, 3]: "list as key"}
except TypeError as e:
    print(e)  # Output: unhashable type: 'list'
    
c. Example of Dictionary Misbehavior with Mutable Keys (If Allowed)
Imagine if Python allowed mutable keys:

# Hypothetical situation where mutable objects are allowed as keys
# If Python allowed this, modifying the key would break the dictionary's lookup
my_dict = {[1, 2]: "value"}

# Modifying the list (which is mutable)
my_key = [1, 2]
my_key.append(3)  # Now my_key is [1, 2, 3]

# Trying to access the dictionary with the modified key would fail
# because the original hash is no longer valid
# (in practice, Python would not allow the list to be a key)
print(my_dict[my_key])  # This would lead to an error if lists were allowed as keys

Conclusion

The immutability of dictionary keys is a fundamental aspect of Python’s dictionary implementation. 
It ensures consistent hashing, efficient lookups, and the integrity of the dictionary data structure. 
Immutable objects like integers, strings, and tuples can be used as dictionary keys, while mutable objects like lists and sets cannot.
This restriction is crucial for maintaining the reliability and performance of dictionaries in Python.
```
